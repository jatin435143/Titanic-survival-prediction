### CALCULATING HOW MUCH ACCURACY A DUMB MODEL THAT JUST GIVES OUT MEAN WILL HAVE

```python
class meanregressor():
    def fit(self,input,target):
        self.mean=target.mean()
    def predict(self,inputs):
        return np.full(inputs.shape[0],self.mean)
     
```

```python
from sklearn.metrics import mean_squared_error as rmse
```

```python
def rmse(pred,target):
    bb=mse(pred,target,squared=False)
    return bb
```

### CREATING LOGISTIC REGRESSION FROM SCRATCH

```python
class logistic_regression:
    
    def __init__(self,lr=0.04,iterations=1000,random_state=None):
        self.lr=lr
        self.random_state = random_state
        self.iterations=iterations
        
    def fit(self,x,y):
        self.x=x
        self.y=y
        rows,columns=self.x.shape
        m=rows
        if self.random_state is not None:
            np.random.seed(self.random_state)
        self.w=np.random.randn(columns)
        self.b=0
        for i in range(self.iterations):
            z=np.dot(self.w,self.x.T)+self.b
            pred=self.sigmoid(z)
            dw=(1/m)*np.dot((pred-self.y),self.x)
            db=(1/m)*np.sum(pred-self.y)
            self.w-=self.lr*dw
            self.b-=self.lr*db
        
    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
        
    def predict(self,x1):
            preds=np.dot(self.w,x1.T)+self.b
            preds=self.sigmoid(preds)
            preds=np.where(preds > 0.5, 1, 0)
            return preds
    def accuracy(pred,target):
        pp=np.array(pred)
        qq=np.array(target).reshape(pp.shape[0])
        acc=((pp==qq).sum()/pp.shape[0])*100
        return acc
```
